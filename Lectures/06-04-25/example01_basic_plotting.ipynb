# **Basic plotting of a sin function**
## Our first exercise
We will use NumPy and Matplotlib to create a series of points and then create a plot

```python
# This is a plotting function
plt.plot(x, y, 'o', color='red')
```

```python
fig, ax = plt.subplots(figsize=(16,10))

x = np.linspace(0,10, 30)
y = np.sin(x)

ax.plot(x,y, 'o', color='red')

x = np.arange(0,10, 0.1)
y = np.sin(x)

ax.plot(x,y, color='blue')
```

```python
fig, ax = plt.subplots(figsize=(16,10))

x = np.linspace(0,10, 30)
y = np.sin(x)

ax.plot(x,y, 'o', color='red', markersize=12)

x = np.arange(0,10, 0.1)
y = np.sin(x)

ax.plot(x,y, color='blue', linestyle='--')

plt.xlim([0.0, 2*np.pi])

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

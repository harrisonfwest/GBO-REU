```python
data_path = '/home/scratch/lmorgan/Projects/GalacticBar/Data/Pilot_All/Pilot_All'
_11_path = data_path + 'Pilot_NH3_11/Pilot_NH3_11_bl2.fits'
_22_path = data_path + 'Pilot_NH3_22/Pilot_NH3_22_bl2.fits'
_33_path = data_path + 'Pilot_NH3_33/Pilot_NH3_33_bl2.fits'
_44_path = data_path + 'Pilot_NH3_44/Pilot_NH3_44_bl2.fits'
_55_path = data_path + 'Pilot_NH3_55/Pilot_NH3_55_bl2.fits'

nu11 = 23.6944955e9 # Hz
nu22 = 23.7239455e9
nu33 = 23.867829e9
nu44 = 24.1369612e9
nu55 = 24.5304249e9
```

```python
# Seeking column density of each species at given points
# Can take (spatial) center value or averaged value for each of the following:
# Main Clump, Shell, Knot A, Knot B, Knot C, Knot D, and Knot E
```

```python
Cube_11 = SpectralCube.read(_11_path).with_spectral_unit(u.km/u.s, velocity_convention='radio', rest_value=nu11*u.Hz)
Cube_22 = SpectralCube.read(_22_path).with_spectral_unit(u.km/u.s, velocity_convention='radio', rest_value=nu22*u.Hz)
Cube_33 = SpectralCube.read(_33_path).with_spectral_unit(u.km/u.s, velocity_convention='radio', rest_value=nu33*u.Hz)
Cube_44 = SpectralCube.read(_44_path).with_spectral_unit(u.km/u.s, velocity_convention='radio', rest_value=nu44*u.Hz)
Cube_55 = SpectralCube.read(_55_path).with_spectral_unit(u.km/u.s, velocity_convention='radio', rest_value=nu55*u.Hz)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 3, figsize = (12, 6))
ax1[0].plot(Cube_11[:, 80, 82])
ax1[0].set_title('(1,1) main clump center')
ax1[0].set_ylim(-.07, 1.02)
# ax1[0].set_xlim(100, 230)
ax1[1].plot(Cube_22[:, 80, 82])
ax1[1].set_title('(2,2) main clump center')
ax1[1].set_ylim(-.07, 1.02)
# ax1[1].set_xlim(100, 230)
ax1[2].plot(Cube_33[:, 80, 82])
ax1[2].set_title('(3,3) main clump center')
ax1[2].set_ylim(-.07, 1.02)
# ax1[2].set_xlim(100, 230)

ax2[0].plot(Cube_44[:, 80, 82])
ax2[0].set_title('(4,4) main clump center')
ax2[0].set_ylim(-.07, 1.02)
# ax2[0].set_xlim(100, 230)
ax2[1].plot(Cube_55[:, 80, 82])
ax2[1].set_title('(5,5) main clump center')
ax2[1].set_ylim(-.07, 1.02)
# ax2[1].set_xlim(100, 230)
fig.show()
```
